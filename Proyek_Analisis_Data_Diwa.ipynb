## Visualization & Explanatory Analysis

```python
# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='season', y='cnt', data=df, estimator=sum, color="red")
plt.title('Total Count per Season')
plt.xlabel('Season')
plt.ylabel('Total Count')
plt.show()
```

```python
# Plot bar chart
plt.figure(figsize=(10, 6))
sns.barplot(x='weathersit', y='cnt', data=df, estimator=sum, color="green")
plt.title('Total Count per weathersit')
plt.xlabel('weathersit')
plt.ylabel('Total Count')
plt.show()
```
